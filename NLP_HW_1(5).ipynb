1. Tokenize a Paragraph   


ఆంధ్రప్రదేశ్‌లో వచ్చే ఎన్నికల నేపథ్యంలో రాజకీయ చర్చలు వేగంగా మారుతున్నాయి. రాష్ట్ర అభివృద్ధి, సంక్షేమ పథకాల అమలు, మరియు ప్రజా సమస్యలపై పార్టీలు తమ తమ అభిప్రాయాలను బలంగా వ్యక్తం చేస్తున్నాయి. ఇటీవల జరిగిన సమావేశాల్లో నాయకులు ప్రజలతో నేరుగా మాట్లాడి తమ ప్రణాళికలను వివరించారు. రాజకీయ వాతావరణం మారుతున్న కొద్దీ ఓటర్ల అంచనాలు కూడా కొత్త దిశలో సాగుతున్నాయి

```python
text = """ ఆంధ్రప్రదేశ్‌లో వచ్చే ఎన్నికల నేపథ్యంలో రాజకీయ చర్చలు వేగంగా మారుతున్నాయి. రాష్ట్ర అభివృద్ధి, సంక్షేమ పథకాల అమలు, మరియు ప్రజా సమస్యలపై పార్టీలు తమ తమ అభిప్రాయాలను బలంగా వ్యక్తం చేస్తున్నాయి. ఇటీవల జరిగిన సమావేశాల్లో నాయకులు ప్రజలతో నేరుగా మాట్లాడి తమ ప్రణాళికలను వివరించారు. రాజకీయ వాతావరణం మారుతున్న కొద్దీ ఓటర్ల అంచనాలు కూడా కొత్త దిశలో సాగుతున్నాయి"""

naive_tokens = text.split()
print(naive_tokens)
```

```python
manual_tokens = [
    'ఆంధ్రప్రదేశ్', 'లో', 'వచ్చే', 'ఎన్నికల', 'నేపథ్యంలో', 'రాజకీయ',
    'చర్చలు', 'వేగంగా', 'మారుతున్నాయి', '.', 'రాష్ట్ర', 'అభివృద్ధి', ',',
    'సంక్షేమ', 'పథకాల', 'అమలు', ',', 'మరియు', 'ప్రజా', 'సమస్యలు', 'పై',
    'పార్టీలు', 'తమ', 'తమ', 'అభిప్రాయాలను', 'బలంగా', 'వ్యక్తం', 'చేస్తున్నాయి', '.',
    'ఇటీవల', 'జరిగిన', 'సమావేశాలు', 'లో', 'నాయకులు', 'ప్రజలతో', 'నేరుగా',
    'మాట్లాడి', 'తమ', 'ప్రణాళికలను', 'వివరించారు', '.', 'రాజకీయ',
    'వాతావరణం', 'మారుతున్న', 'కొద్దీ', 'ఓటర్ల', 'అంచనాలు', 'కూడా', 'కొత్త',
    'దిశలో', 'సాగుతున్నాయి'
]

print(manual_tokens)
```

2. Comparison of Tokenization Methods    

Let's compare the `naive_tokens` (simple split), `manual_tokens`, and `indic_tokens`.

```python
print("Naive Tokens:", naive_tokens)
print("Manual Tokens:", manual_tokens)
print("Indic NLP Tokens:", indic_tokens)

# Compare naive and Indic NLP token sets (symmetric difference)
diff_naive_indic = set(naive_tokens) ^ set(indic_tokens)
print("\nDifferences between Naive and Indic NLP Tokens (symmetric difference):")
print(diff_naive_indic)

# Compare manual and Indic NLP token sets (symmetric difference)
diff_manual_indic = set(manual_tokens) ^ set(indic_tokens)
print("\nDifferences between Manual and Indic NLP Tokens (symmetric difference):")
print(diff_manual_indic)

# Show directional differences for clearer inspection
print("\nTokens in Naive but not in Indic NLP:", set(naive_tokens) - set(indic_tokens))
print("Tokens in Indic NLP but not in Naive:", set(indic_tokens) - set(naive_tokens))

print("\nTokens in Manual but not in Indic NLP:", set(manual_tokens) - set(indic_tokens))
print("Tokens in Indic NLP but not in Manual:", set(indic_tokens) - set(manual_tokens))
```

### Analysis of Differences
Naïve tokenization simply splits on spaces, so words like "మారుతున్నాయి.", "అభివృద్ధి,", and "సాగుతున్నాయి." keep their punctuation attached. Manual tokenization removes this punctuation and also splits meaningful suffixes, turning "ఆంధ్రప్రదేశ్‌లో" into "ఆంధ్రప్రదేశ్" + "లో", which is more linguistically accurate. The Indic NLP tool separates punctuation correctly (e.g., ".", ",") but does not split suffixes, so it keeps "ఆంధ్రప్రదేశ్‌లో" as a single token. Because Telugu is agglutinative, tool-based tokenization often misses internal boundaries that humans can identify. Overall, naïve tokenization struggles with punctuation, Indic NLP struggles with morphology, and manual tokenization produces the cleanest Telugu‑appropriate tokens.


### 3. Multiword Expressions (MWEs)

Examples from the Telugu paragraph:
- ఆంధ్రప్రదేశ్‌లో వచ్చే ఎన్నికలు
This phrase refers to a specific political event. If we split it word‑by‑word, the meaning becomes weaker, so it works best as one semantic unit.
- సంక్షేమ పథకాల అమలు
These words together describe a single policy action. Separating them breaks the idea of “implementation of welfare schemes,” which is understood as one concept.
- రాజకీయ వాతావరణం
This is an idiomatic expression meaning “political climate.” The meaning is not just “politics + atmosphere,” so it should be treated as one meaningful phrase.


### 4. Reflection
Tokenizing Telugu turned out to be challenging mainly because the language is agglutinative, and words like "ఆంధ్రప్రదేశ్‌లో" or "సాగుతున్నాయి." often contain multiple suffixes and attached punctuation. Unlike English, where splitting by spaces works most of the time, Telugu requires more linguistic awareness to separate meaningful parts correctly. Punctuation frequently sticks to the end of words, and removing it without losing meaning takes extra care. Morphology also plays a big role, since a single Telugu word can express what would take several words in English. Multiword expressions such as "రాజకీయ వాతావరణం" or "సంక్షేమ పథకాల అమలు" add another layer of difficulty because they function as single ideas even though they contain multiple words. Overall, Telugu tokenization is more complex than English due to its structure, suffixes, and semantic phrases that need to be preserved
